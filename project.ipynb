```python
# Dfinir une fonction pour extraire les caractristiques audio
def extract_features(file_path):
    # Charger le fichier audio avec Librosa
    y, sr = lr.load(file_path, sr=None)
    
    # Calculer le spectre ml
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr)
    
    # Calculer le chromagramme
    chroma = librosa.feature.chroma_stft(y=y, sr=sr)
    
    # Calculer le rythme
    tempo, _ = librosa.beat.beat_track(y=y, sr=sr)

    return mel_spec, chroma, tempo
```

```python
import os

# Dfinir les motions
EMOTIONS = {1: 'neutral', 2: 'calm', 3: 'happy', 4: 'sad', 5: 'angry', 6: 'fear', 7: 'disgust', 8: 'surprise'}

# Chemin du dossier contenant les fichiers audio
DATA_PATH = fr'C:\Users\doria\OneDrive\Bureau\ECE\ING5\DL\projet_dl\projet_DL\dataset'

# Dfinir la liste pour stocker les donnes
data = []

# Parcourir les dossiers
folders = glob.glob(os.path.join(DATA_PATH, '*'))
for folder in folders:
    # Parcourir les fichiers dans chaque dossier
    folder1 = glob.glob(os.path.join(folder, '*'))
    for files1 in folder1:
        data1 = glob.glob(os.path.join(files1, '*'))
        for file_path in data1:
            # Charger le fichier audio avec Librosa
            y, sr = lr.load(file_path, sr=None)
            
            # Extraire les informations du fichier
            identifiers = os.path.basename(file_path).split('.')[0].split('-')
            
            modality = int(identifiers[0])
            vocal_channel = int(identifiers[1])
            emotion = int(identifiers[2])
            emotion_intensity = int(identifiers[3])
            statement = int(identifiers[4])
            repetition = int(identifiers[5])
            actor = int(identifiers[6])
            
            # Mapping des informations pour correspondre  votre structure
            emotion_mapping = {1: 'neutral', 2: 'calm', 3: 'happy', 4: 'sad', 5: 'angry', 6: 'fear', 7: 'disgust', 8: 'surprise'}
            emotion = emotion_mapping[emotion]
            emotion_intensity = 'normal' if emotion_intensity == 1 else 'strong'
            gender = 'female' if actor % 2 == 0 else 'male'
            
            # Ajouter les informations au DataFrame
            data.append({
                'Modality': modality,
                'Vocal Channel': vocal_channel,
                'Emotion': emotion,
                'Emotion Intensity': emotion_intensity,
                'Statement': statement,
                'Repetition': repetition,
                'Actor': actor,
                'Gender': gender,
                'Path': file_path
            })

# Convertir la liste en DataFrame
df = pd.DataFrame(data)

# Afficher le DataFrame
print(df)
```

```python
for index, row in df.iterrows():
    file_path = row['Path']
    mel_spec, chroma, tempo = extract_features(file_path)

    # Stockez les caractristiques rsumes dans le DataFrame
    df.at[index, 'Mel Spectrogram'] = mel_spec.mean()  # Moyenne pour rduire la dimentionnalit
    df.at[index, 'Chroma'] = chroma.mean()            
    df.at[index, 'Tempo'] = tempo
```

```python
# Normaliser les caractristiques
scaler = StandardScaler()
df[['Mel Spectrogram', 'Chroma', 'Tempo']] = scaler.fit_transform(df[['Mel Spectrogram', 'Chroma', 'Tempo']])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv1D, MaxPooling1D
from keras.utils import to_categorical

# Convertir les tiquettes d'motion en encodage one-hot
le = LabelEncoder()
df['Emotion'] = le.fit_transform(df['Emotion'])
y = to_categorical(df['Emotion'], len(EMOTIONS))

# Sparer les donnes en ensembles d'entranement et de test
X_train, X_test, y_train, y_test = train_test_split(df[['Mel Spectrogram', 'Chroma', 'Tempo']], y, test_size=0.2, random_state=42)
X_train = X_train.values.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test = X_test.values.reshape((X_test.shape[0], X_test.shape[1], 1))

# Dfinir le modle
model = Sequential()

# Ajouter une couche Convolution 1D avec MaxPooling
model.add(Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(X_train.shape[1],)))
model.add(MaxPooling1D(pool_size=2))

# Ajouter une deuxime couche Convolution 1D avec MaxPooling
model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
model.add(MaxPooling1D(pool_size=2))

# Aplatir les caractristiques pour les passer  travers les couches entirement connectes
model.add(Flatten())

# Ajouter les couches entirement connectes
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dense(len(EMOTIONS), activation='softmax'))

# Compiler le modle
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Entraner le modle
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2, verbose=1)

# Prdire sur l'ensemble de test
y_pred = model.predict(X_test)
y_pred_labels = np.argmax(y_pred, axis=1)
y_test_labels = np.argmax(y_test, axis=1)

# valuer les performances
accuracy = accuracy_score(y_test_labels, y_pred_labels)
conf_mat = confusion_matrix(y_test_labels, y_pred_labels)
class_report = classification_report(y_test_labels, y_pred_labels)

print(f'Accuracy: {accuracy}')
print(f'Confusion Matrix:\n{conf_mat}')
print(f'Classification Report:\n{class_report}')
```
